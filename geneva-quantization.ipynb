```python
import jax.numpy as np
from jax import grad, vmap
import matplotlib.pyplot as plt
```

```python
# Model for 4-external straight-slot wheel
def geneva4(theta):
    return np.arctan(np.sin(theta)/(np.sqrt(2)-np.cos(theta)))
def sigmoid(x):
    return np.exp(x)/(np.exp(x)+1)
```

```python
thetas = np.linspace(-np.pi/4, np.pi/4, 50)
```

```python
plt.plot(x, geneva4(thetas)+np.pi/4)
plt.plot(x, 1.5*sigmoid(thetas*10))
```

```python
# manual gradient, since JAX hasn't implemented atan gradient
def grad_geneva4(theta):
    f = lambda t: np.sin(t)/(np.sqrt(2)-np.cos(t))
    return grad(f)(theta)/(1+f(theta)**2)
```

```python
plt.plot(thetas, vmap(grad_geneva4)(thetas))
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, vmap(grad(sigmoid))(x))
```
